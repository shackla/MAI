```
from keras import models
import os
import pylab
import numpy as np
from skimage import io
import csv
from google.colab import drive
```

```
drive.mount('/content/drive')

parent_dir = '/content/drive/My Drive/Colab Notebooks/Testing/Specs/Ml'
model_dir = 'good models/Custms1.h5'
img_dir = ['test/direct_speech', 'test/bg', 'test/bab']
results_dir = ['results/cust_cat3_test_ds.csv', 'results/cust_cat3_test_env.csv', 'results/cust_cat3_test_bab.csv']
tags = ['-direct.jpg', '-bg.jpg', '_test.jpg']
```

```
model_path = os.path.join(parent_dir, model_dir)
```

```
model = models.load_model(model_path)
print(model_path)
```

```
for j in range(0, 3):
  results_path = os.path.join(parent_dir, results_dir[j])
  image_path = os.path.join(parent_dir, img_dir[j])
  for i in range(3000, 12000):
      filename = str(i) + tags[j]
      filepath = os.path.join(image_path, filename)
      if os.path.exists(filepath):
        all_images = []
        img = io.imread(filepath)
        img = img.reshape([224, 224, 3])
        img = img/225
        all_images.append(img)
        x = np.array(all_images)
        answer = model.predict(x, batch_size=None, verbose=1, steps=None)
        print(answer)


        row = [filename, answer]

        with open(results_path, 'a') as csvFile:
            writer = csv.writer(csvFile)
            writer.writerow(row)

        csvFile.close()
        print('Written')
      else: 
        print('FileNotFound')
```
