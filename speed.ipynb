```
from keras import models
import os
import pylab
import librosa
import librosa.display
import numpy as np
from PIL import Image
from skimage import io
import csv
import time
from google.colab import drive
```

```
drive.mount('/content/drive')
parent_dir = '/content/drive/My Drive/Colab Notebooks/Testing'
audio_dir = "test"
image_dir = "img"
model_file = ['models/custbin.h5', 'models/resbin.h5']
csv_file = "results/results_cnns_w_imgpros"
```

```
audio_path = os.path.join(parent_dir, audio_dir)
image_path = os.path.join(parent_dir, image_dir)
csv_path = os.path.join(parent_dir, csv_file)
```

```
dimension = 224


max_val = [1, 10, 100, 500]
```

```

for i in range(100):
    start = time.time()
    chunkname = str(i+1) + ".wav"
    audio_file = os.path.join(audio_path, chunkname)

    save_path = os.path.join(image_path, str(i)+"control.jpeg")

    sig, fs = librosa.load(audio_file)
    pylab.axis('off')
    pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[])
    S = librosa.feature.melspectrogram(y=sig, sr=fs)
    librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
    pylab.savefig(save_path, bbox_inches=None, pad_inches=0)
    pylab.close()
    im = Image.open(save_path)
    resize = im.resize((dimension, dimension), Image.ANTIALIAS)
    resize.save(save_path)

    all_images = []
    img = io.imread(save_path)
    img = img.reshape([dimension, dimension, 3])
    img = img/225
    all_images.append(img)
    x = np.array(all_images)

    end = time.time()
    dur = end-start
    print(dur)

        
```
