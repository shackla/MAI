```
drive.mount('/content/drive')

parent_dir = '/content/drive/My Drive/Colab Notebooks/Testing/live'
large_dir = 'mp3'
sub_dirs = ['ds', 'bab', 'babc', 'bg']
save_dir = 'spilt'
tags = ['-ds', '-bg', '-bab']
```

```
base_path = os.path.join(parent_dir,large_dir)

save_path = os.path.join(parent_dir, save_dir)
```

```
dur = 3000
filename = os.path.join(base_path, 'direct_speech_live.mp3')
sound = AudioSegment.from_mp3(filename)

num_spilts = int(round(len(sound)/dur)+1)
print(num_spilts)
for i in range(1, num_spilts):
  x = dur*(i-1)
  y = dur*(i)
  z = i+7
  section = sound[x:y]
  chunkname = str(z) +tags[0] +'.wav'
  save_file = os.path.join(save_path, sub_dirs[0],chunkname)
  section.export(save_file, format="wav")
  print('spilt number '+str(z)+' is complete')
```
